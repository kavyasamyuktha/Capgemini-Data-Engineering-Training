```sql
select e.emp_id, e.emp_name, sum(s.amount) as total_sales
from employee e join sales s
on e.emp_id = s.emp_id
group by e.emp_name, e.emp_id;
```

```sql
SELECT e.emp_id, e.emp_name, COUNT(s.sales_id) AS sales_count
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_id, e.emp_name;
```

```sql
SELECT s.product, SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY s.product;
```

```sql
SELECT s.product, AVG(s.amount) AS avg_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY s.product;
```

```sql
SELECT e.emp_id, e.emp_name, COUNT(s.sales_id) AS sales_count
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_id, e.emp_name
HAVING COUNT(s.sales_id) > 2;
```

```sql
select emp_id, sum(amount) as total_sales
from sales
group by emp_id;
```

```sql
select emp_id, count(distinct product) as unique_products
from sales
group by emp_id;
```

```sql
select emp_id, max(amount) as max_sales
from sales
group by emp_id;
```

```sql
select product, sum(amount) as total_sales
from sales
group by product
having sum(amount) > 50000;
```

```sql
select e.department, avg(s.amount) from employee e 
join sales s on
e.emp_id = s.emp_id
group by e.department
order by avg(s.amount)
limit 1;
```

```sql
select e.department, max(s.amount) from employee e 
join sales s on
e.emp_id = s.emp_id
group by e.department
order by max(s.amount)
limit 1;
```

```sql
select e.emp_name, sum(s.amount) from employee e
join sales s on
e.emp_id = s.emp_id
group by e.emp_name
order by sum(s.amount) desc
limit 3;
```

```sql
SELECT YEAR(joining_date) AS year,
       COUNT(*) AS emp_count,
       AVG(salary) AS avg_salary
FROM Employee
GROUP BY YEAR(joining_date);
```

```sql
select e.department, sum(s.amount) as total_sales
from employee e
join sales s on
e.emp_id = s.emp_id
group by e.department;
```

```sql
select e.department, count(e.emp_id) AS no_sales_count
from Employee e
join Sales s ON e.emp_id = s.emp_id
where s.emp_id IS NULL
group by e.department;
```
