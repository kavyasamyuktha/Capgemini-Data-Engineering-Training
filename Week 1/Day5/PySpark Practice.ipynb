```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.getOrCreate()
data = [
    (1, "Alice", 1000),
    (2, "Bob", 2000),
    (3, "Charlie", 3000),
    (4, "David", None),
    (5, "Eva", 1500)
]
```

```python
df = spark.createDataFrame(data, ["id", "name", "salary"])
```

```python
df.show()
```

```python
df.display()
```

```python
df.withColumn("bonus", col("Salary")*0.1).display()
```

```python
df.drop(("id")).display()
```

```python
df.withColumnRenamed("name", "employee_name").display() 
```

```python
df.withColumn("double_salary", col("salary")*2).display()
```

```python
df.withColumnRenamed("name", "employee_name").display()
```

```python
df.select("name", "salary").show()
```

```python
df.filter(col("salary") > 1500).show()
```

```python
df.withColumn("salary_str", col("salary").cast("string")).show()
```

```python
df.select(col("salary").alias("sal")).show()
```

```python
df.orderBy(col("salary").desc()).show()
```

```python
df.orderBy("salary", ascending=False).show()
```

```python
df.orderBy("salary", ascending=0).show()
```

```python
df.orderBy(["salary", "id"], ascending=[False, True]).show()
```

```python
df.sort("salary").show()
```

```python
df.display()
```

```python
from pyspark.sql.functions import lit

df = df.withColumn("country", lit("India"))
df.display()
```

```python
data2 = [
    (1, "Alice Johnson"),
    (2, "Bob Smith"),
    (3, "Charlie Brown"),
    (4, "David Miller"),
    (5, "Eva Taylor")
]

df_names = spark.createDataFrame(data2, ["id", "full_name"])

print("=== Full Name Table ===")
df_names.display()
```

```python
from pyspark.sql.functions import split, col

df_split = df_names.withColumn("first_name", split(col("full_name"), " ")[0]) \
                   .withColumn("last_name", split(col("full_name"), " ")[1])

df_split.display()
```
