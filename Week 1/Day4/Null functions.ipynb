```sql
select emp_id, name, ifnull(salary, 0) as salary
from Employees;
```

```sql
select emp_id, name, ifnull(bonus, 1000) as bonus from employees;
```

```sql
select order_id, ifnull(amount, 500) as amount from orders;
```

```sql
select product_id, product_name, ifnull(stock, 0) as stock from products;
```

```sql
select emp_id, name,
       coalesce(salary, bonus) as earnings
from Employees;
```

```sql
select emp_id, name,
       coalesce(salary, bonus, 0) as earnings
from Employees;
```

```sql
select product_id, product_name,
       coalesce(price, 0) as price
from products;
```

```sql
select order_id,
       coalesce(amount, discount, 0) as payment
from Orders;
```

```sql
select emp_id, nullif(salary, 0) as salary from employees;
```

```sql
select order_id, nullif(discount, 0) as discount from orders;
```

```sql
select order_id,
       nullif(coupon_code, 'DISC10') as coupon_code
from Orders;
```

```sql
select emp_id, name, ifnull(salary, 0) + ifnull(bonus, 0) as total_earning from Employees;
```

```sql
select * from employees where salary is null and bonus is null;
```

```sql
select * from orders where amount is null and discount is null;
```

```sql
select emp_id, coalesce(salary, bonus, 1000) as income from employees;
```

```sql
select order_id, nullif(discount, 0) as discount from orders;
```

```sql
select order_id, ifnull(amount, 0) - ifnull(discount, 0) as final_payabale_amount from orders;
```
